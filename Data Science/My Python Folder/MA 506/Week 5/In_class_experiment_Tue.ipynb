```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.01,10,100)
y = np.sin(x)*3 + np.random.normal(size = [100,])
plt.scatter(x,y)
plt.show()
```

$Y = \beta_0 + \beta_1x + \beta_2x^2$

```python
n = 100
p = 3
```

```python
X = np.array([[i**0,i**1,i**2] for i in x])
beta = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
beta
```

```python
ypred = X.dot(beta)
## plot
plt.scatter(x,y)
plt.plot(x,ypred,color = 'r')
plt.show()
```

```python
e = y-ypred
sigmahat = np.sqrt(np.linalg.norm(e)**2/(n-p))
sigmahat
```

```python
H = X.dot(np.linalg.inv(X.T.dot(X))).dot(X.T)
H.shape
```

```python
diagH = np.diag(H)
diagH.shape
```

```python
eistar = e/(sigmahat*np.sqrt(1-diagH))
eistar.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.scatter(x,eistar)
plt.axhline(y = -2,color = 'r')
plt.axhline(y = 2,color = 'r')
plt.axhline(y = -3,color = 'k')
plt.axhline(y = 3,color = 'k')
plt.show()
```

```python
d = (eistar**2*diagH)/(p*(1 - diagH))
d.shape
```

```python
plt.scatter(x,d)
plt.show()
```
