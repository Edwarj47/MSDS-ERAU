```python
import numpy as np
import matplotlib.pylab as plt
```

```python

def df(n):
    res = 1;
    for i in range(n, -1, -2):
        if(i == 0 or i ==1):
            return res;
        else:
            res *= i;
print("Double Factorial is: ", df(int(input())))
```
